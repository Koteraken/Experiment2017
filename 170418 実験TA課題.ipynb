```python
url3 = 'http://archive.ics.uci.edu/ml/machine-learning-databases/ecoli/ecoli.data'
urllib.urlretrieve(url3, 'ecoli.txt')
#df3 = pd.read_csv('ecoli.txt', sep=' ', index_col=0) 
```

```python
name = ["mcg", "gvh", "lip", "chg", "aac", "alm1", "alm2", "cp", "pp", "im", "imS", "imL", "imU", "om", "omL"]
li = []
for line in open("ecoli.txt", "r"):
    a = line.strip().split()
    #print a
    ali = []
    #print line 
    for aa in a:
        try:
            ali.append(float(aa))
            
        except:
            
            if aa == 'cp':
                ali.extend([0, 0, 0, 0, 0, 0, 0, 0])
                ali[-8] = int(1)
            if aa == 'pp':
                ali.extend([0, 0, 0, 0, 0, 0, 0, 0])
                ali[-7] = int(1)
                continue
            if aa == 'im':
                ali.extend([0, 0, 0, 0, 0, 0, 0, 0])
                ali[-6] = int(1)
                continue
            if aa == 'imS':
                ali.extend([0, 0, 0, 0, 0, 0, 0, 0])
                ali[-5] = int(1)
                continue
            if aa == 'imL':
                ali.extend([0, 0, 0, 0, 0, 0, 0, 0])
                ali[-4] = int(1)
                continue 
            if aa == 'imU':
                ali.extend([0, 0, 0, 0, 0, 0, 0, 0])
                ali[-3] = int(1)
                continue
            if aa == 'om':
                ali.extend([0, 0, 0, 0, 0, 0, 0, 0])
                ali[-2] = int(1)
                continue
            if aa == 'omL':
                ali.extend([0, 0, 0, 0, 0, 0, 0, 0])
                ali[-1] = int(1)
                continue 
    ali = map(float, ali[:])            
    li.append(ali) 
```
